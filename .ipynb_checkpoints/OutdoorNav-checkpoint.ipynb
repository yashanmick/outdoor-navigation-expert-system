```python
import sqlite3
```

```python
nav_db = 'outdoorNav.sqlite'
```

```python
conn = sqlite3.connect(nav_db)
cur = conn.cursor()
SQL = "select * from NAVIGATION"
cur.execute(SQL)
```

```python
#convert the table "NAVIGATION" to pandas dataframe
import pandas as pd

data = cur.fetchall()

colnames = [desc[0] for desc in cur.description]

df = pd.DataFrame(data)
df.columns = colnames
```

```python
df
```

```python
#calculate road width
import numpy as np

df['RoadWidth'] = (df['StartWidth'] + df['EndWidth']) / 2
```

```python
df
```

```python
#loop through the df
#format each row in the following format,
#and export it to txt file ("facts.txt") as the input to CLIPS
#
#(nav
#    (object-width 0.66)
#    (object-depth-from-begining 3.16)
#    ((object-to-left-boarder 1.06)
#    (object-to-right-boarder 2.25)
#    (road-width3.315)
#    (frame-point 0.6)
#    )
```

```python
fout = open('D:/Projects/OutdoorNavigation/facts.txt', 'w')

fout.write("(initial-fact)\n")

for index, row in df.iterrows():
    line = "(nav (object-width %g) (object-depth-from-begining %g) (object-to-left-boarder %g) (object-to-right-boarder %g) (road-width %g) (frame-point %g))\n" \
    %(row['ObjectWidth'],
     row['ObjectDepthFromBegining'],
     row['ObjectToLeftBoarder'],
     row['ObjectToRightBoarder'],
     row['RoadWidth'],
     row['FramePointer'])
    
    fout.write(line)
    
fout.close()
```
